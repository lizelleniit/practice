```python
import gym
import numpy as np
import gym_windy_gridworlds
env = gym.make('WindyGridWorld-v0')

env.reset()

gam=1
eps=0.1
alpha=0.5
done = False
policy=np.zeros((7,10))
Q=np.zeros((7,10,4))
numeps=170
for ep in range(numeps):
    s=(3,0)
    env.observation=s
    oldpolicy=policy.copy() # save the old policy
    policy=np.argmax(Q,axis=2) # update the policy

    ### pick an action according to the policy ###
    if np.random.random()<eps:
        a=np.random.randint(4)
    else:
        a = policy[s[0],s[1]]#env.action_space.sample()
    ###################################################
    done=False
    while not done:
        ##################################################
        if ep==numeps-1: # print out the path for the last episode
            print("s: ",s,"\ta: ",a)
            env.render(mode='human', close=False)
        ###################################################
        
        sp, r, done, info = env.step(a) # take action a, observe r, sp
        env.observation=sp
        policy=np.argmax(Q,axis=2) # update the policy according to Q

        ### pick an action according to the policy ###
        if np.random.random()<eps:
            ap=np.random.randint(4)
        else:
            ap=policy[sp[0],sp[1]]
        ##############################################
        Q[s[0],s[1],a]=Q[s[0],s[1],a]+alpha*(r+gam*Q[sp[0],sp[1],ap]-Q[s[0],s[1],a]) # update Q
        
        a=ap
        s=sp
```
